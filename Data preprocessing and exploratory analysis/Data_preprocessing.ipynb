# ***Importing the data***

```python
import pandas as pd

df=pd.read_excel("/content/drive/MyDrive/Dissertation/My Dissertation/Data/Bitcoin price and other currency rates in USD.xlsx")
df.tail(21)
```

# ***Arrange the data in cronological order***

```python
df=df.loc[::-1].reset_index(drop=True)
df.head(21)
```

#  ***Handling missing values***



**Count the total number of missing values for each column and display it**

```python
count = df.isna().sum()
NaN_count=pd.DataFrame(count,columns=['missing_count'])
NaN_count.sort_values(by='missing_count', ascending=False) 
```

**Fill Friday's close value for Saturday and Sunday**

```python
final_df=df
column_list=final_df.columns

for i in column_list:
  final_df[i]=final_df[i].fillna(method='ffill')

final_df.head(20)
```

**Recheck the total number of missing values for each column and display it**

```python
count = final_df.isna().sum()
NaN_count=pd.DataFrame(count,columns=['missing_count'])
NaN_count.sort_values(by='missing_count', ascending=False) 
```

**Check the data types of each column**

```python

print(final_df.dtypes)
```
